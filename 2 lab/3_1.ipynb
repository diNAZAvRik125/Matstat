```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint, norm, kstest, cauchy
import pandas as pd
import seaborn as sns
```

```python
excel_data = pd.read_excel('Коновалов Назар.xlsx')
samples_1 = pd.DataFrame(excel_data, columns=['sample'])
samples = np.array(samples_1[['sample']])
samples = samples.ravel()
```

```python
tetta = np.arange(np.size(samples), dtype=np.double)
s = 0
for i in range(np.size(samples)):
    s+=samples[i]
    tetta[i] = s/(i+1)
```

```python
tetta_g = tetta[2000::]
plt.plot(np.arange(0, np.size(tetta_g)), tetta_g)
plt.axhline(-5, color='red', linestyle='--')
plt.show()
```

```python
tetta_1 = np.arange(10**3, dtype=np.float64)
D = np.arange(10**3, dtype=np.float64)
print(np.size(tetta_1))
```

```python
O = -5
f = 0
for i in range(10**3):
    s = samples[i:(10**3+i)]
    f = np.sqrt(np.mean(s**2) - (np.mean(s))**2)
    tetta_1[i] = (np.sqrt(10**3)*(np.mean(s) - O))
    D[i] = f
```

```python
print(np.sqrt(np.mean(tetta_1**2) - (np.mean(tetta_1))**2))
```

```python
x = np.linspace(-5, 5, np.size(tetta_1))
plt.hist(tetta_1, density = True, alpha = 0.3, bins = int(len(tetta_1)**(2/5)))
plt.plot(x, norm.pdf(x, 0, 1.87), 'k--', label='Standard normal')
# plt.legend()
plt.show()
```

```python
x = np.sort(tetta_1)
y = np.arange(1, np.size(x)+1) / np.size(x)
x_norm = np.linspace(min(tetta_1), max(tetta_1), np.size(tetta_1))
y_norm = norm.cdf(x_norm, 0, 1.87)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='Empirical Distribution Function')
ax.plot(x_norm, y_norm, 'k--', label='Normal Distribution')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('Cumulative Probability')
plt.show()
```
