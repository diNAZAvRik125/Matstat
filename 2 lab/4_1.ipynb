```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint, norm, kstest, cauchy
import pandas as pd
import seaborn as sns
```

```python
excel_data = pd.read_excel('Коновалов Назар.xlsx')
samples_1 = pd.DataFrame(excel_data, columns=['sample'])
samples = np.array(samples_1[['sample']])
samples = samples.ravel()
```

### АНО для нормального распредиение это мат ожидание(из лекций Черновой) ###

```python
tetta = np.mean(samples)
sigma = np.sqrt(np.mean(samples**2) - (np.mean(samples))**2)
```

```python
eps = (1+0.99)/2
kvant_2 = (norm.isf(1-eps))
print(kvant_2)
```

```python
eps = (1+0.95)/2
kvant_1 = (norm.isf(1-eps))
print(kvant_1)
```

```python
x = np.linspace(-28, 28, np.size(samples))
level_line = -5
level_lower = tetta-(kvant_1*sigma/np.sqrt(np.size(samples)))
level_upper = tetta+(kvant_1*sigma/np.sqrt(np.size(samples)))

# Define the confidence interval as a shaded region
plt.fill_between(x, level_lower, level_upper, alpha=0.2)

# Plot the data and the level line
# plt.plot(x, y, label='Data')
plt.axhline(y=level_line, color='red', linestyle='--', label='Level Line')

# Add a legend and axis labels
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
