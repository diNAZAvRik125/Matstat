```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint, norm, kstest, cauchy
import pandas as pd
import seaborn as sns
```

```python
excel_data = pd.read_excel(' .xlsx')
samples_1 = pd.DataFrame(excel_data, columns=['sample'])
```

```python
samples = np.array(samples_1[['sample']])
```

```python
samples = samples.ravel()
```

```python
b = max(samples)
a = min(samples)
print(a, b)
```

```python
a = np.mean(samples)
sigma = np.sqrt(np.mean(samples**2) - a**2)
print(a, sigma)
```

```python
x = np.linspace(-28, 28, np.size(samples))
plt.hist(samples, histtype='step', density = True, bins=int(len(samples)**(2/5)))
plt.plot(x, norm.pdf(x, a, sigma), 'k--', label='Standard normal')
plt.show()
```

```python
sample_norm = norm.rvs(a, sigma, size = 1000000, random_state=0)
```

```python
kstest(samples, sample_norm)
```
