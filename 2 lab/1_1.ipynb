```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint, norm, kstest, cauchy
import pandas as pd
import seaborn as sns
```

```python
excel_data = pd.read_excel(' .xlsx')
samples_1 = pd.DataFrame(excel_data, columns=['sample'])
```

```python
print(samples_1)
```

```python
samples = np.array(samples_1[['sample']])
print(samples)
```

```python
samples = samples.ravel()
```

```python
samples
```

```python
b = max(samples)
a = min(samples)
print(a, b)
```

```python
std_norm = norm()
```

```python
x = np.linspace(-28, 28, np.size(samples))
plt.hist(samples, histtype='step', density = True, bins=int(len(samples)**(2/5)))
plt.plot(x, norm.pdf(x, -5, 4.8), 'k--', label='Standard normal')
plt.show()
```

```python
x = np.sort(samples)
y = np.arange(1, np.size(x)+1) / np.size(x)
x_norm = np.linspace(min(samples), max(samples), np.size(samples))
y_norm = norm.cdf(x_norm, -5, 4.8)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='Empirical Distribution Function')
ax.plot(x_norm, y_norm, 'k--', label='Normal Distribution')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('Cumulative Probability')
plt.show()
```
