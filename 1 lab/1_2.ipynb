```python
import numpy as np
from numpy.random import default_rng
from scipy import stats
from bokeh.plotting import figure, show
from scipy.stats import chisquare
from scipy.stats import randint
```

```python
def generation(n):
    xk_g = stats.uniform.rvs(loc = 3, scale = 7, size = n)
    return(xk_g)
```

```python
def S_n(xk, n):
    s = 0
    for i in range(n):
        s += xk[i]
    return s/n
```

```python
def E_xk(xk):
    return np.mean(xk)
```

```python
xk = stats.uniform.rvs(loc = 3, scale = 7, size = 10)
Sn = np.float32(np.arange(16))
Exk = np.float32(np.arange(16))
delta = np.float32(np.arange(16))
Exk[0] = E_xk(xk)
Sn[0] = S_n(xk, 10)
delta[0] = float(1000*(S_n(xk, 10) - E_xk(xk)))
xk = generation(25)
Exk[1] = E_xk(xk)
Sn[1] = S_n(xk, 25)
xk = generation(50)
Exk[2] = E_xk(xk)
Sn[2] = S_n(xk, 50)
xk = generation(75)
Exk[3] = E_xk(xk)
Sn[3] = S_n(xk, 75)
xk = generation(100)
Exk[4] = E_xk(xk)
Sn[4] = S_n(xk, 100)
xk = generation(250)
Exk[5] = E_xk(xk)
Sn[5] = S_n(xk, 250)
xk = generation(500)
Exk[6] = E_xk(xk)
Sn[6] = S_n(xk, 500)
xk = generation(750)
Exk[7] = E_xk(xk)
Sn[7] = S_n(xk, 750)
xk = generation(1000)
Exk[8] = E_xk(xk)
Sn[8] = S_n(xk, 1000)
xk = generation(2500)
Exk[9] = E_xk(xk)
Sn[9] = S_n(xk, 2500)
xk = generation(5000)
Exk[10] = E_xk(xk)
Sn[10] = S_n(xk, 5000)
xk = generation(7500)
Exk[11] = E_xk(xk)
Sn[11] = S_n(xk, 7500)
xk = generation(10000)
Exk[12] = E_xk(xk)
Sn[12] = S_n(xk, 10000)
xk = generation(100000)
Exk[13] = E_xk(xk)
Sn[13] = S_n(xk, 100000)
xk = generation(1000000)
Exk[14] = E_xk(xk)
Sn[14] = S_n(xk, 1000000)
xk = generation(10000000)
Exk[15] = E_xk(xk)
Sn[15] = S_n(xk, 10000000)
```

```python
N = np.linspace(0, 16, 16)
E = []
for i in range(16):
    E.append(6.5)
p_1 = figure(title="Multiple glyphs example", x_axis_label="x", y_axis_label="y")
p_1.line(N, Sn, legend_label="Temp.", color="red", line_width=2)
p_1.line(N, E, legend_label="Rate", color="blue", line_width=2)
show(p_1)
```
