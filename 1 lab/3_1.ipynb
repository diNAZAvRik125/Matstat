```python
import numpy as np
from numpy.random import default_rng
from scipy import stats
from bokeh.plotting import figure, show
from scipy.stats import chisquare
from scipy.stats import randint
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
T = np.linspace(0, np.pi/4, 6)
print(T)
```

```python
def n(xk, t):
    sn = 0
    n = 0
    xk = np.sort(xk)
    while((sn < t) and (n < np.size(xk))):
        sn += xk[n]
        n += 1
    return n
```

```python
def generator(mu,N):
    xk_10 = stats.poisson.rvs(mu, size = 10)
    xk_100 = stats.poisson.rvs(mu, size = 100)
    xk_1000 = stats.poisson.rvs(mu, size = 1000)
    xk_10000 = stats.poisson.rvs(mu, size = 10000)
    xk_100000 = stats.poisson.rvs(mu, size = 100000)
    xk_1000000 = stats.poisson.rvs(mu, size = 1000000)
    N.append(n(xk_10, mu))
    N.append(n(xk_100, mu))
    N.append(n(xk_1000, mu))
    N.append(n(xk_10000, mu))
    N.append(n(xk_100000, mu))
    N.append(n(xk_1000000, mu))
    return N
```

```python
T = np.linspace(0, np.pi/4, 60)
N = []
for i in range(10):
    N = generator(T[i], N)
print(N)
```

```python
x = np.linspace(0, 12, 1000)
y = stats.poisson.pmf(x, np.pi/4)
plt.plot (x, y, color='red', linewidth= 3)
count = plt.(N, 0, T, colors='k', linestyles='-', lw=1,
        label='frozen pmf')
plt.show()
```

```python
# Prepare data
N = 1000 # step
lambdas = [0, 1/2,  0.78539816]
X_T = [np.random.poisson(lam, size=N) for lam in lambdas]
S = [[np.sum(X[0:i]) for i in range(N)] for X in X_T]
X = np.linspace(0, N, N)
 
# Plot the graph
graphs = [plt.step(X, S[i], label="Lambda = %d"%lambdas[i])[0] for i in range(len(lambdas))]
plt.legend(handles=graphs, loc=2)
plt.title("Poisson Process", fontdict={'fontname': 'Times New Roman', 'fontsize': 21}, y=1.03)
plt.ylim(0)
plt.xlim(0)

lambdas_1 = [1/2]
X_T = [np.random.poisson(1/2, size=N)]
S = [[np.sum(X[0:i]) for i in range(N)] for X in X_T]
X = np.linspace(0, N, N)
graphs = [plt.step(X, S[i], label="Lambda_1 = %d"%lambdas_1[i])[0] for i in range(len(lambdas_1))]
plt.show()
```
