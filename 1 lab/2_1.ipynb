```python
E = (10+3)/2
print(E)
```

```python
def generation(n):
    xk_g = stats.uniform.rvs(loc = 3, scale = 7, size = n)
    return(xk_g)
```

```python
def S_n(xk, n):
    s = 0
    for i in range(n):
        s += xk[i]
    return s
```

```python
def f(n, CPT, xk):
    s = 0
    d = stats.uniform.var(3, 10)
    for i in range(n):
        s += xk[i]
        a = ((s-(i+1)*E)/np.sqrt((i+1)*d))
        CPT.append(a)
    return CPT
```

```python
xk = stats.uniform.rvs(loc = 3, scale = 7, size = 10)
CPT = []
CPT = f(10, CPT, xk)
xk = generation(25)
CPT = f(25, CPT, xk)
print(CPT)
xk = generation(50)
CPT = f(50, CPT, xk)
xk = generation(75)
CPT = f(75, CPT, xk)
xk = generation(100)
CPT = f(100, CPT, xk)
xk = generation(250)
CPT = f(250, CPT, xk)
xk = generation(500)
CPT = f(500, CPT, xk)
xk = generation(750)
CPT = f(750, CPT, xk)
xk = generation(1000)
CPT = f(1000, CPT, xk)
xk = generation(2500)
CPT = f(2500, CPT, xk)
xk = generation(5000)
CPT = f(5000, CPT, xk)
xk = generation(7500)
CPT = f(7500, CPT, xk)
xk = generation(10000)
CPT = f(10000, CPT, xk)
```

```python
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 2, 1)

ax1.hist(CPT, color = 'blue', edgecolor = 'black', density=True,
         bins = int(150))
x = np.arange (-4, 4, 0.0001)
ax1.plot (x, norm.pdf(x, 0, 1), color='red', linewidth= 3)
 
plt.show()
```
