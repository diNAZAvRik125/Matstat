```python
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import pareto
from scipy.stats import norm
import seaborn as sns
import numpy as np
 
values = []
values_for_imperiar = []
 
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
 
r = pareto(6, 2)
 
mv = r.moment(1)
dv = scipy.stats.pareto.var(6, 2)
 
values = [(sum(pareto.rvs(6, 2, size=i)) - (mv * i))/((i*dv)**0.5) for i in range(50, 50000, 5)]
# for i in range(10, 50000, 5):
#     e = pareto.rvs(6, 2, size=i)
#     s = sum(e)
#     values.append((s - (mv * i))/((i*dv)**0.5))
 
ax1.hist(values, color = 'blue', edgecolor = 'black', density=False,
         bins = int(150))
 
x = np.arange (-4, 4, 0.001)
ax1.plot (x, norm.pdf(x, 0, 1), color='red', linewidth= 3)
 
for i in range(1000):
    e = pareto.rvs(6, 2, size=200)
    s = sum(e)
    values_for_imperiar.append((s - (mv * 200)) / ((200 * dv) ** 0.5))
 
ax2 = sns.ecdfplot(values_for_imperiar)
x = np.linspace(-4,4,100)
norm_rv = norm(0, 1)
cdf = norm_rv.cdf(x)
ax2.plot(x, cdf)
 
plt.show()
```
